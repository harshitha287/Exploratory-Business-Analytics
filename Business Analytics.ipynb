## Exploratory Data analysis

### Correlation

```python
corr = df.corr()
sns.heatmap(corr,annot=True,cmap='Greens')
```

### Visualizing the dataset

```python
sns.pairplot(df,height=2);
```

### Visualizing individual parameters

```python
df.hist(figsize=(13,13),bins=50)
plt.show()
```

#### 1.Most Customers tends to buy quantity of 2 and 3.

#### 2.Discount given maximum is 0 to 20 percent

```python
#visualizing Ship mode
print(df['Ship Mode'].value_counts())
sns.countplot(x = df["Ship Mode"])
plt.show()
```

#### Observation:'Standard Class' is the preferred shipping mode whereas 'Same Day' is the least preferred shipping mode

```python
#visualizing Segment
print(df['Segment'].value_counts())
sns.countplot(x = df['Segment'],palette="rocket_r")
plt.show()
```

#### Highest segment is of "Consumer" and Lowest segment is of "Home office"

```python
#Visualizing Region
theme = plt.get_cmap('hsv')
print(df['Region'].value_counts())
plt.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.title('Percentage of sales in diff Regions of USA')
```

#### Maximum sales were in "West Region" and Minimum sales were in "South Region"

```python
#Visualizing Category
print(df['Category'].value_counts())
plt.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,autopct='%1.1f%%')
plt.title('Percentage of sales for different categories')
```

#### Most selling category is "Office supllies" and Least selling category is "Technolgy"

```python
plt.figure(figsize=(10,8))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.2f%%')
plt.show()
```

#### Most of the items from Blinders,Paper,furnishings,phones,storage,arts,Accessories and least from copiers,machines,supplies

```python
#Visualizing Sub-category
print(df['Sub-Category'].value_counts())
sns.countplot(x = df['Sub-Category'])
plt.xticks(rotation = 90)
plt.show()
```

#### "Binders" are high in demand whereas "Copiers" are not of use to many

```python
#Visualizing State
plt.figure(figsize=(15,5))
sns.countplot(x = df.State)
plt.xticks(rotation = 90)
plt.show()
```

#### maximum number of people are from "California" and "New York" whereas least number of people are from "Wyoming"

### Business Analysis

#### Profit, Loss and Sales visualization

```python
#State wise analysis of Profit and Sales:
df.groupby('State')[['Profit', 'Sales']].sum().plot.bar(color = ["red", "blue"],figsize= (18,8))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

#### California has maximum sales with good profit whereas many states such as Texas and Ohio face loss

```python
#City wise analysis of Profits and sale
print(df['City'].value_counts())
City = df.groupby("City")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False)
City[:30].plot(kind = "bar", color = ["black", "gray"],figsize= (18,8))
plt.title("Profit, Loss & Sales of top 30 Cities")
plt.xlabel("Cities")
plt.ylabel("Total profit / loss and sales")
plt.show()
```

#### New york has maximum sales and profit. Philadelphia, Houston, Chicago, Jacksonville, San antonio, Burlington, Dallas, Charlotte, Louisville and Denver face loss

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','purple'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

#### Profit and Sales are maximum in Consumer Segment and minimum in Home Office Segment.

```python
#Category wise analysis of Profit and Sales
df.groupby('Category')[['Profit', 'Sales']].sum().plot.bar(color = ["yellow", "orange"],figsize= (8,5))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

#### Although sales of Office supplies is not higher than Furniture but Office supplies make more profit

```python
#Region wise analysis of profit and sales
df.groupby('Region')[['Profit', 'Sales']].sum().plot.bar(color = ["cyan", "lightgreen"],figsize= (8,5))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

#### Sales and profit are high in West region
